```python
import pandas as pd
from numpy.polynomial.polynomial import Polynomial
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import UnivariateSpline, interp1d, lagrange
```

```python
df = pd.read_csv("v_avgvar2500.txt", sep=" ", header=None)
df1 = df.iloc[:, [0,1]]
```

```python
f = interp1d(df1[0], df1[1])
spl = UnivariateSpline(df1[0], df1[1])

x_new = np.arange(0, 100, 0.1)
plt.scatter(df[0], df[1], label='data')
plt.plot(x_new, f(x_new), label='Interp1d', color='orange')
plt.plot(x_new, spl(x_new), label='UnivariateSpline', color='green')
plt.legend()
plt.show()
```
